```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_excel('datasets\Big_retail_dataset.xlsx')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']).dt.normalize()
print("Количество строк: %d"  %  df.shape[0])
df.head()
```

```python
gp_InvoiceDate = df.groupby('InvoiceDate')
transactions = []
for name,group in gp_InvoiceDate:
    transactions.append(list(group['Description'].map(str)))

print(len(transactions))


transactions_df=pd.DataFrame(transactions)
transactions_df.head()
```

```python
association_rules = apriori(transactions, min_support = 0.1, min_confidence = 0.75, min_lift = 3, max_length = 2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
for item in association_results:

    pair = item[0]
    items = [x for x in pair]

    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
last_transaction=transactions[len(transactions)-1]
print(len(last_transaction))
print(len(set(last_transaction)))
print(last_transaction)
```

```python
future_transaction_list=[]
for item in association_results:

    pair = item[0]
    items = [x for x in pair]

    for transact_item in last_transaction:
        if(items[0]==transact_item):
            future_transaction_list.append(items[1])

print(len(future_transaction_list))
print(future_transaction_list)

future_transaction=set(future_transaction_list)

print(len(future_transaction))
print(future_transaction)
```
